```python
import nltk
from nltk import word_tokenize , pos_tag
nltk.download('tagsets_json')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
import pandas as pd
```

```python
from nltk.help import brown_tagset
print(brown_tagset())
```

```python
text1 = word_tokenize("And now for something completely different")
pos = nltk.pos_tag(text1)
text2= word_tokenize("Artificial intelligence is the science and engineering of making machines think like humans")
pos = nltk.pos_tag(text2)
for word, tag in pos:
    print(f"Word: {word}, Tag: {tag}")
    print(brown_tagset(tag))
```

```python
df=pd.read_csv('/content/wiki_movie_plots_deduped.csv.zip')
df.head()
```

```python
data = df.Plot
for data in data:
  sentences=nltk.sent_tokenize(data)
for sentence in sentences:
  words=nltk.word_tokenize(sentence)
  pos=nltk.pos_tag(words)
print(pos)
```
